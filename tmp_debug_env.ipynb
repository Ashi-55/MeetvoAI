# Debugging .env.local and Gemini Model Selection

This notebook inspects the `.env.local` file, verifies the Gemini model value, and updates it if needed.

```python
from pathlib import Path

path = Path('c:/Users/DELL/OneDrive/Desktop/MeetvoAI/.env.local')
text = path.read_text(encoding='utf-8')
print('EXACT CONTENTS OF .env.local:')
print(repr(text))

text = text.replace('GOOGLE_AI_MODEL=gemini-1.5-flash', 'GOOGLE_AI_MODEL=gemini-1.0')
path.write_text(text, encoding='utf-8')
print('\nUpdated .env.local successfully.')
print(repr(path.read_text(encoding='utf-8')))
```
